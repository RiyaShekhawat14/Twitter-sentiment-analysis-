```python
### import all the dependencies using in the model 
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
###stopwords are the words that do not enter any meaning to the data 
import nltk
nltk.download("stopwords")
stop_words=stopwords.words("english")
```

```python
### Data preprocessing
twitter_data=pd.read_csv("training.1600000.processed.noemoticon.csv",encoding="latin-1")
```

```python
### for checking rows and columns of data
twitter_data.shape
```

```python
#### printing first 5 rows of the dataset
twitter_data.head()
```

```python
###naming the columns and reading the dataset again
column_names=['target','id','date','flag','user','text']

twitter_data=pd.read_csv("training.1600000.processed.noemoticon.csv",names=column_names ,encoding="latin-1")

twitter_data.head()
```

```python
twitter_data.shape
```

```python
twitter_data.isnull().sum()
```

```python
###checking the distribution of target 
twitter_data["target"].value_counts()
```

```python
### convert the target "4" to "1"
twitter_data.replace({'target':{4:1}}, inplace=True)
twitter_data.head()
twitter_data["target"].value_counts()
```

```python
#### 0---->negative tweet
#### 1--->poisitve tweet
```

```python
port_stem=PorterStemmer()
```

#### Stemming--Stemming is the process of reducing a word to its Root word
port_stem=PorterStemmer()   

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stop_words]
    stemmed_content = ' '.join(stemmed_content)

    return stemmed_content
```

```python

twitter_data['stemmed_content']=twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data["stemmed_content"])
```

```python
twitter_data.shape
```

```python
print(twitter_data['target'])
```

```python
###Separting the data and target
x=twitter_data['stemmed_content'].values
y=twitter_data['target'].values
```

```python
print(x)
```

```python
print(y)
```

```python
####Splitting the data to trainijng data and test data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, stratify=y ,random_state=2)    
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_test)
```

```python
print(x_train)
```

```python
###converting textual data to numerical data 

vectorizer=TfidfVectorizer()

x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
####Training the machine learning model - Logistic Regression
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
### Model evualtion
```

```python
####ACcuracy score on the training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score on the training data:',training_data_accuracy)
```

```python
####ACcuracy score on the test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score on the testing data:',testing_data_accuracy)
```
